```python
part = dict()
for record in records:
    if (record.find(' грн.') != -1) and (record.find(' 0.00 грн.') == -1):
        part[record.split(',')[0].lower()] = float(record[(record.rfind('-')+2):record.rfind(' грн.')])
part
```

```python
capital = sum(part.values())
capital
```
